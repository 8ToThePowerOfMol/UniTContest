## Testování hypotézy

Statistiku budu testovat pomocí testu korelace normálního rozdělení.

### Koeficient korelace

Koeficient korelace vektoru $(X_1,Y_1), ... , (X_n,Y_n)$ odhadnu pomocí výběrového koeficientu korelace $r_{xy}$ vypočteného jednoprůchodovým vzorcem:
$$r_{xy}=\frac{n\sum_{j=1}^{n}x_jy_j-(\sum_{j=1}^{n}x_j)(\sum_{j=1}^{n}y_j)}{\sqrt{(n\sum_{j=1}^{n}x^2_j-(\sum_{j=1}^{n}x_j)^2)(n\sum_{j=1}^{n}y^2_j-(\sum_{j=1}^{n}y_j)^2))}}=\frac{n}{n-1}\frac{\frac{1}{n}\sum_{j=1}^{n}x_jy_j-\overline{\textbf{x}}\overline{\textbf{y}}}{s_{\textbf{x}}s_{\textbf{y}}}$$
kde $\overline{\textbf{x}}$ je výběrový průměr náhodné veličiny $X$, $\overline{\textbf{y}}$ je výběrový průmer náhodné veličiny $Y$, $s_\textbf{x}$ je výběrová směrodatná odchylka (corrected sample standard deviation) náhodné veličiny $X$ a $s_\textbf{x}$ je výběrová směrodatná odchylka náhodné veličiny $Y$.

```python
def get_correlation_coeficient(x : str, y : str, df : pd.DataFrame):
    n = len(df)
    sum_xy = (df[x] * df[y]).sum()
    return (sum_xy - n * df[x].mean() * df[y].mean()) / ((n-1) * df[x].std() * df[y].std())
```

#### Poznámka ke statistickým funkcním v pythonu
K výpočtu průměru a odchylky využívám funkcí knihovny `pandas`:
- [pd.DataFrame.sum()](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.sum.html) vrací součet datového rozsahu.
- [pd.DataFrame.mean()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.mean.html) vrací průměr datového rozsahu, v našem případě se jedná výběrový průměr.
- [pd.DataFrame.std()](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.std.html) vrací směrodatnou odchylku. Běžně se jedná o výběrovou směrodatnou odchylku upravenou Besselovou korekcí s jedním stupněm volnosti. Funkci lze přetížit pomocí atributu `ddof` na požadovaný stupeň volnosti, takže např. `df[x].std(ddof=0)` by nám vrátilo směrodatnou odchylku (uncorrected sample standard deviation) náhodné veličiny $X$.

### Testovací statistika

Testovací statistika odvozena od výběrového koeficientu korelace je:
$$t=\frac{r_{\textbf{x},\textbf{y}}\sqrt{n-2}}{\sqrt{1-r_{\textbf{x},\textbf{y}}^2}}$$

Tuto statistku budu testovat na Studentově rozdělení o $n-2 = 148$ stupních volnosti.
Kritická hodnota statisky se určí jako kvantil ze Studentova rozdělení s hladinou významnsti $\alpha$, tedy
$$q_{t(n-2)}(1-\frac{\alpha}{2})=q_{t(148)}(0.975)$$
Kvantil Studentova rozdělení vyjde

```python
st.t.ppf(0.975, 148)
```

tedy pokud vyjde $| t |>q_{t(148)}(0.975)=1.9761224936033632$, hypotézu zamítáme.

```python
def is_statistics_accepted(t : float, significance, degree):
    return abs(t) > st.t.ppf(1 - significance/2, degree-2)

def get_statistics(x : str, y : str, df : pd.DataFrame):
    r = get_correlation_coeficient(x, y, df)
    n = len(df)
    return r * math.sqrt(n-2) / math.sqrt(1-r**2)

def is_correlating(x : str, y : str, df : pd.DataFrame, significance):
    return is_statistics_accepted(get_statistics(x, y, df), significance, len(df))
```

Statistika pro $X$ a $Y_{Happiness Score}$ vyjde například takto:

```python
get_statistics('GDP per Capita', 'Happiness Score', df)
```

## Výsledky korelace HDP a jednotlivých faktorů štěstí

```python
# null hypothesis: x, y are correlating
print(is_correlating('GDP per Capita', 'Happiness Score', df, 0.05))
print(is_correlating('GDP per Capita', 'Economy (GDP per Capita)', df, 0.05))
print(is_correlating('GDP per Capita', 'Family', df, 0.05))
print(is_correlating('GDP per Capita', 'Health (Life Expectancy)', df, 0.05))
print(is_correlating('GDP per Capita', 'Freedom', df, 0.05))
print(is_correlating('GDP per Capita', 'Trust (Government Corruption)', df, 0.05))
print(is_correlating('GDP per Capita', 'Generosity', df, 0.05))
print(is_correlating('GDP per Capita', 'Dystopia Residual', df, 0.05))
```

### Kritická mez

Pro určení kritické meze (tj. meze, nad níž zamítáme nulovou hypotézu) vyjdu z určení statistiky podle hledaného koeficientu korelace: $$q_{t(n-2)}(1-\frac{\alpha}{2})=q_{t(148)}(0.975)>t=\frac{r_{\textbf{x},\textbf{y}}\sqrt{n-2}}{\sqrt{1-r_{\textbf{x},\textbf{y}}^2}}$$
Nechť prozatím je $q:=q_{t(148)}(0.975)$. Hledám $k$, aby $$q>\frac{k\sqrt{148}}{\sqrt{1-k^2}}$$
po umocnění
$$q^2(1-k^2)>148k^2$$
$$q^2-q^2k^2>148k^2$$
$$q^2>148k^2+q^2k^2$$
a $k$ vyjde
$$k<\sqrt{\frac{q^2}{148-q^2}}$$
spočítám:

```python
math.sqrt(st.t.ppf(0.975, 148)**2 / (148-st.t.ppf(0.975, 148)**2))
```

To potvrzuje, že pokud výběrový korelační koeficient (naše koeficinety jsou vypočítány v heatmapě výše) vyjde nižší, než 0.16462261516917115, nulovou hypotézu zamítáme.

## *Dystopia Residual*
*Dystopia Residual* lze podle [popisu dat v Kaggle](https://www.kaggle.com/unsdsn/world-happiness/home) chápat jako nevysvětlitelný antiutopický zbytek, který má negativní podíl na štěstí. Státy, ve kterých funguje totalitní režim, nebo se uvnitř těchto států válčí budou mít *Dystopia Residual* vyšší, než státy žijící v míru. Jak to vypadá s rozvojovými státy už není snadné vidět, proto se zde více podívám na tuto složku.

Setřídím státy podle *Dystopia Residual*:

```python
df_float = copy.deepcopy(df)
for col in df_float.loc[:, df_float.columns != 'Country']:
    df_float[col].apply(lambda x: float(x))
    
df_float.sort_values(by='Dystopia Residual', ascending=0)
```

Ještě vykreslím HDP a *Dystopia Residual*:

```python
sorted_dystopia_series = df_float.sort_values(by='Dystopia Residual', ascending=0)['Dystopia Residual'].reset_index(drop=True)
plt.figure(figsize=(12, 10), dpi=60)
plt.plot(sorted_dystopia_series, 'o')
```

(V každém grafu matplotlib je zde vodorovná osa pořadí státu a svislá osa vykreslovaná náhodná veličina)

```python
sorted_gdp_series = df_float.sort_values(by='GDP per Capita', ascending=0)['GDP per Capita'].reset_index(drop=True)
plt.figure(figsize=(12, 10), dpi=60)
plt.plot(sorted_gdp_series, 'o')
```

Zde je vidět, že *Dystopia Residual* má normální rozdělení (což přepokládm podle toho, že data jsou získávána z lidské populace), ale HDP jistě nemá normální rozdělení. Za předpokladu, že jsme určovali nulovou hypotézu korelace dvou veličin normálního rozdělelní, bylo by chybné prohlásit toto za konečný výsledek.

Platnost mé hypotézy, že data *Dystopia Residual* pochází z normálního rozdělení, ale HDP už ne rychle ověřím testem pomocí [scipy.stat.normaltest()](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.normaltest.html#scipy.stats.normaltest):

```python
def has_normal_distribution(data, significance_alpha):
    k, p = st.normaltest(data)
    # null hypothesis: data comes from a normal distribution
    if p < significance_alpha:
        return False, p
    else:
        return True, p

print(has_normal_distribution(df['GDP per Capita'], 0.05))
print(has_normal_distribution(df['Dystopia Residual'], 0.05))
```

## Test s logaritmickonormálním rozdělením

Nyní zlogarimuji všechna numerická data. Protože hodnoty HDP jsou vysoké (Lucembursko ohromně vysoké), zmenší se HDP hodně, ale *Happiness Score* a další jeho faktory štěstí se příliž nezmenší. Protože se všehny veličiny zmenší úměrně svému logaritmu, neočekávám, že jejich rozdíly nevlivní příliž jejich korelaci.

```python
df_log = copy.deepcopy(df_float)
for col in df_log.loc[:, df_log.columns != 'Country']:
    df_log[col] = df_log[col].apply(lambda x: math.log(x) if x != 0 else np.NAN)

df_log
```

Setříděné HDP a *Dystopia Residual* po logaritmování vypadají takto:

```python
sorted_dystopia_series = df_log.sort_values(by='Dystopia Residual', ascending=0)['Dystopia Residual'].reset_index(drop=True)
plt.figure(figsize=(12, 10), dpi=60)
plt.plot(sorted_dystopia_series, 'o')
```

```python
sorted_gdp_series = df_log.sort_values(by='GDP per Capita', ascending=0)['GDP per Capita'].reset_index(drop=True)
plt.figure(figsize=(12, 10), dpi=60)
plt.plot(sorted_gdp_series, 'o')
```

To ještě není zcela normální rozdělení. Nicméně HDP už se více "blíží" normálnímu rozdělení.

Korelační matice vypadá takto:

```python
df_sub = df_log[['GDP per Capita',\
             'Happiness Score',\
             'Economy (GDP per Capita)',\
             'Family','Health (Life Expectancy)',\
             'Freedom',\
             'Generosity',\
             'Trust (Government Corruption)',\
             'Dystopia Residual']]
corr_matrix = df_sub.corr()
corr_matrix
```

```python
sns.set(rc={'figure.figsize':(8,8)})
sns.heatmap(corr_matrix, square=True, cmap="YlGnBu", annot=True, fmt=".2f")
```

Korelační matice zde už vyšla odlišně, než korelační matice normálního rozdělení. Zde je vidět, že došlo k velkému snížení korelace mezi HDP a *Generosity* a to až o přibližně desetkrát. Naopak nejvíce korelující *Happiness Score* a *Economy* se s HDP shodují více, což bylo očekáváno.

Ještě je potřeba vzít v úvahu i logaritmování nuly, které není definováno. Pokud v datech logartimuji 0, výsledek nahradím np.NAN, což bude potom hodnota ignorovaná.

Nyní znovu použiji už předem napsané funkce k určení testu hypotézy. Tentokrát, podle korelační matice očekávám, že zamítnutých hypotéz bude více.

```python
print(is_correlating('GDP per Capita', 'Happiness Score', df_log, 0.05))
print(is_correlating('GDP per Capita', 'Economy (GDP per Capita)', df_log, 0.05))
print(is_correlating('GDP per Capita', 'Family', df_log, 0.05))
print(is_correlating('GDP per Capita', 'Health (Life Expectancy)', df_log, 0.05))
print(is_correlating('GDP per Capita', 'Freedom', df_log, 0.05))
print(is_correlating('GDP per Capita', 'Trust (Government Corruption)', df_log, 0.05))
print(is_correlating('GDP per Capita', 'Generosity', df_log, 0.05))
print(is_correlating('GDP per Capita', 'Dystopia Residual', df_log, 0.05))
```

# Výsledek

U testu korelace s normálním rozdělením jsme museli zamítnout pouze nulovou hypotézu korelace mezi $X$ a $Y_{Dystopia Residual}$. U testu korelace kde jsem zlogaritmoval všechna data musíme navíc zamítnout i korelaci mezi $X$ a $Y_{Generosity}$. Ostatní hypotézy závislosti HDP a štěstí nebo nějakého faktoru štěstí obyvatel státu můžeme přijmout.
